```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
## Definizione di intervalli e funzioni

dati_1 = input('Inserisci xmin, xmax: ')
xmin, xmax = dati_1.split(',')
x = np.linspace(eval(xmin),eval(xmax),200)

dati_2 = input('Inserisci ymin, ymax: ')
ymin, ymax = dati_2.split(',')

functions = input('Inserisci le funzioni da disegnare: ').split(',')
```

```python

fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')

for f in functions:
    ax.plot(x,eval(f), label=f.split('np.')[-1])
    
ax.set_ylim([eval(ymin),eval(ymax)])  
ax.grid()
ax.legend();
```
